```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('hairpin_counts.csv')
data = data.set_index(data.columns[0])
```

```python
data.head()
```

```python
print('S3 Máximo:'+ str(np.max(data.iloc[0].values)))
print('S3 Mínimo:'+ str(np.min(data.iloc[0].values)))
print('S1 Máximo:'+ str(np.max(data.iloc[1].values)))
print('S1 Mínimo:'+ str(np.min(data.iloc[1].values)))
print('S2 Máximo:'+ str(np.max(data.iloc[2].values)))
print('S2 Mínimo:'+ str(np.min(data.iloc[2].values)))
```

```python
print('S3 Total:'+ str(sum(data.iloc[0].values)))
print('S1 Total:'+ str(sum(data.iloc[1].values)))
print('S2 Total:'+ str(sum(data.iloc[2].values)))
```

```python
data = data + 1
datalog=np.log10(data)
```

```python
datalog.head()
```

```python
row_index = 0
row_values = datalog.iloc[row_index].values
plt.hist(row_values, bins=6)
plt.title(f'Sample 3 miRNA Counts')
plt.xlabel('log10(Counts)')
plt.ylabel('Frequency')
plt.show()
```
